```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load CSV file
df = pd.read_csv(r"Final_File")
df
```

```python
df = df.drop(columns=[col for col in ["Unnamed: 0.1", "Unnamed: 0"] if col in df.columns])
```

```python
df1 = df.copy()
df1
```

```python
# Display first few rows
print(df.head())
```

```python
# Check dataset information (columns, data types, missing values)
print(df.info())
```

```python
# Check summary statistics (for numeric columns)
print(df.describe())
```

```python
# Count missing values in each column
print(df.isnull().sum())
```

```python
print(df.isna().sum().sum())  # If 0, there are no NaN values
```

```python
# Find duplicate rows
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert text to lowercase
df["Brand"] = df["Brand"].str.lower()

# Remove spaces
df["Brand"] = df["Brand"].str.strip()

# Fix inconsistent spellings (e.g., ‘samsung’ vs ‘Samsung’)
df["Brand"] = df["Brand"].replace({"samsng": "samsung", "realme ": "realme"})
```

```python
df
```

```python
# Convert price, storage, and RAM to integer
df["Price"] = df["Price"].astype(int)
df["Storage"] = df["Storage"].astype(int)
df["RAM"] = df["RAM"].astype(int)
```

```python
df.info()
```

```python
# Remove non-tablet brands
df = df[~df["Brand"].str.contains("dell", case=False, na=False)]

# Remove rows with unrealistic values
df = df[df["Price"] > 1000]  # Keep only products above ₹1000
```

```python
df
```

```python
df["Processor"]
```

```python
Processor = []

for i in df['Processor']:
    Processor.append(re.sub(r'Processor:\s*', '', i) if re.search(r'Processor:\s*', i) else np.nan)
```

```python
df['Processor'] = df['Processor'].str.replace(r'Processor:\s*', '', regex=True)
```

```python
df["Battery"]
```

```python
df['Battery'] = df['Battery'].str.extract(r'Battery:\s*(\d+)', expand=False)
```

```python
df['Feature']
```

```python
df['Feature'] = df['Feature'].str.extract(r'(1\s+Year)', expand=False)
```

```python
df['Feature'].dropna()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df
```

```python
df['Camera']
```

```python
df[['Rear_Camera', 'Front_Camera']] = df['Camera'].str.extract(r'(\d+\.?\d*\s*MP).*?\|\s*(\d+\.?\d*\s*MP)', expand=True)
```

```python
df.head()
```

```python
df = df.drop(columns= "Camera")
```

```python
df.info()
```

```python
df["Rear_Camera"] = df["Rear_Camera"].fillna(df["Rear_Camera"].mode()[0])
df["Front_Camera"] = df["Front_Camera"].fillna(df["Front_Camera"].mode()[0])
```

```python
df[['Rear_Camera', 'Front_Camera']]
```

```python
df['Rear_Camera'] = df['Rear_Camera'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['Front_Camera'] = df['Front_Camera'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
```
